```python
import pandas as pd
from sklearn import linear_model

def model_data(X, y):
    model = linear_model.LinearRegression()\
       .fit(X, y)
    r2 = model.score(X, y)
    coef = model.coef_[0]
    return model, r2, coef
    
```

```python
df = pd.DataFrame([
    (2, 3.28512382507),
    (3, 3.5569589138),
    (4, 4.74807214737),
    (5, 6.92266011238),
    (6, 9.49729299545),
    (7, 15.121505022),
    (8, 19.9062268734)
], columns=["Variables", "Secs"])

df["Classes"] = 2 ** df["Variables"]

model, r2, coef = model_data(df[["Classes"]], df["Secs"])
print "R^2:", r2
print "%f ms for every case class" % (1000. * coef)

xlim = [0, 300.]
df.plot(kind="scatter", x="Classes", y="Secs", xlim=xlim)
plot(xlim, model.predict(np.array([xlim]).T))
```

```python
implicits = pd.DataFrame([
    (10, 3.94771099091),
    (20, 4.02056384087),
    (30, 4.10808706284),
    (40, 4.19162583351),
], columns=["Implicits", "Secs"])

_, r2, coef = model_data(implicits[["Implicits"]], implicits["Secs"])
print "R^2:", r2
# there are 20 invocations in each test
print "%f ms per implicit per function invocation" % (1000. * coef / 20.)

implicits.set_index("Implicits").plot(kind="bar", ylim=[3.8, 4.2])
```
